```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# read file
u_cma, u_truth = np.load(os.path.join('diffusion_cma_60s.npy'))
u_sgd, _ = np.load(os.path.join('diffusion_sgd_minibatch_60s.npy'))
```

```python
np.mean((u_cma - u_truth)**2), np.mean((u_sgd - u_truth)**2)
```

```python
# plot solution
#plt.rc('font', family='serif')
#plt.rc('xtick', labelsize='medium')
#plt.rc('ytick', labelsize='medium')
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

x = np.linspace(0, 1, 10000)

# visualization (solution)
fig = plt.figure(figsize=(6, 3.5))
plt.plot(x, u_truth, linewidth=4, alpha=.5, c='cyan', label = 'Ground truth');
plt.plot(x, u_truth, linewidth=4, alpha=.0, c='cyan', label = ' ');
plt.plot(x, u_cma, linewidth=1.5, alpha=1, c='darkgreen', linestyle='-', label = 'CMA-ES');
plt.plot(x, u_sgd, linewidth=1.5, alpha=1, c='r', linestyle='dashed', label = 'SGD');
plt.legend(ncol=2, frameon=False, fontsize='large'); plt.xlabel('x', fontsize='large'); plt.ylabel('u', fontsize='large');
```
