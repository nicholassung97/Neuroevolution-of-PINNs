```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# read file
x_cma, y_cma, x_truth, y_truth = np.load('projectile_cma_60s.npy')
x_sgd, y_sgd, _, _ = np.load('projectile_sgd_minibatch_60s.npy')
```

```python
np.mean((np.hstack([x_truth,y_truth])-np.hstack([x_cma,y_cma]))**2), np.mean((np.hstack([x_truth,y_truth])-np.hstack([x_sgd,y_sgd]))**2)
```

```python
# plot solution
#plt.rc('font', family='serif')
#plt.rc('xtick', labelsize='medium')
#plt.rc('ytick', labelsize='medium')
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

plist = np.arange(0, 10000, 500)
plist = np.append(plist, 10000-1)

# visualization (solution)
fig = plt.figure(figsize=(6, 3.5))
plt.plot(x_truth, y_truth, linewidth=3, alpha=.5, c='cyan', label = 'Ground truth');
plt.plot(x_truth, y_truth, linewidth=3, alpha=.0, c='cyan', label = ' ');
plt.plot(x_cma, y_cma, linewidth=1.5, alpha=1, c='darkgreen', linestyle='-', label = 'CMA-ES');
plt.plot(x_sgd, y_sgd, linewidth=1.5, alpha=1, c='r', linestyle='dashed', label = 'SGD');
plt.legend(ncol=2, frameon=False, fontsize='large'); plt.xlabel('x', fontsize='large'); plt.ylabel('y', fontsize='large');
```
