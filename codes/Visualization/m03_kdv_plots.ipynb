```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# read file
u_cma, u_truth = np.load('kdv_cma_60s.npy')
u_sgd, _ = np.load('kdv_sgd_minibatch_60s.npy')
```

```python
np.mean((u_cma - u_truth)**2), np.mean((u_sgd - u_truth)**2)
```

```python
# plot solution
#plt.rc('font', family='serif')
#plt.rc('xtick', labelsize='medium')
#plt.rc('ytick', labelsize='medium')
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

# visualization (solution)
fig = plt.figure(figsize=(20, 3))
plt.subplots_adjust(bottom=0, left=0, top=1, right=1, wspace=.2);

con_lv = 25

# plot boundary 
x_l, x_u = 0, 1.5
t_T = 2.0
ext = [x_l, x_u, 0 ,t_T]    # plot boundary 

# ground truth
ax = fig.add_subplot(1, 5, 1)
ax = plt.imshow(u_truth.T, interpolation='bilinear', origin='lower', cmap='rainbow', vmin=0, vmax=1, extent=ext, aspect='auto');
plt.xticks(np.arange(-0, 2., step=.5)); plt.yticks(np.arange(0, 2.5, step=.5));
plt.colorbar(); plt.xlabel('x', fontsize='large'); plt.ylabel('t', fontsize='large'); plt.title('Ground truth', fontsize='x-large'); 

ax = fig.add_subplot(1, 5, 2)
ax = plt.imshow(u_cma.T, interpolation='bilinear', origin='lower', cmap='rainbow', vmin=0, vmax=1, extent=ext, aspect='auto');
plt.xticks(np.arange(-0, 2., step=.5)); plt.yticks(np.arange(0, 2.5, step=.5));
plt.colorbar(); plt.title('CMA-ES', fontsize='x-large');

ax = fig.add_subplot(1, 5, 3)
ax = plt.imshow(abs(u_truth - u_cma).T, interpolation='bilinear', origin='lower', cmap='hot', vmin=0, vmax=0.25, extent=ext, aspect='auto');
plt.xticks(np.arange(-0, 2., step=.5)); plt.yticks(np.arange(0, 2.5, step=.5));
plt.colorbar(); plt.title('abs. err', fontsize='x-large');

ax = fig.add_subplot(1, 5, 4)
ax = plt.imshow(u_sgd.T, interpolation='bilinear', origin='lower', cmap='rainbow', vmin=0, vmax=1, extent=ext, aspect='auto');
plt.xticks(np.arange(-0, 2., step=.5)); plt.yticks(np.arange(0, 2.5, step=.5));
plt.colorbar(); plt.title('SGD', fontsize='x-large');

ax = fig.add_subplot(1, 5, 5)
ax = plt.imshow(abs(u_truth - u_sgd).T, interpolation='bilinear', origin='lower', cmap='hot', vmin=0, vmax=0.25, extent=ext, aspect='auto');
plt.xticks(np.arange(-0, 2., step=.5)); plt.yticks(np.arange(0, 2.5, step=.5));
plt.colorbar(); plt.title('abs. err', fontsize='x-large');
```
